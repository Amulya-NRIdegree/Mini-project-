```python
df['Quantity']=pd.to_numeric(df['Quantity'])
df['Price']=pd.to_numeric(df['Price'])
df['Discount']=pd.to_numeric(df['Discount'])

df['Discount']= df['Discount'].fillna(0)
```

```python
df.dropna(subset=['Quantity','Price'], inplace=True)
```

```python
df['TransactionDate'] = pd.to_datetime(df['TransactionDate'])
```

```python
df['NetPrice'] = df['Price'] * (1 - df['Discount'] / 100)
```

```python
df['TotalSale'] = df['Quantity'] * df['NetPrice']
```

```python
total_revenue = df['TotalSale'].sum()
print(f"Total Revenue: {total_revenue}")
```

```python
most_frequent_product = df['ProductID'].mode()[0]
print(f"Most Frequently Purchased Product: {most_frequent_product}")
```

```python
average_transaction_amount = df['TotalSale'].mean()
print(f"Average Transaction Amount: {average_transaction_amount}")
```

```python
high_value_transactions = df[df['TotalSale'] > 200]
print("Transactions with Total Amount Greater than 200:")
print(high_value_transactions)
```
