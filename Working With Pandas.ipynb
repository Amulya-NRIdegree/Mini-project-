```python
df['Quantity']=pd.to_numeric(df['Quantity'])
df['Price']=pd.to_numeric(df['Price'])
df['Discount']=pd.to_numeric(df['Discount'])

df['Discount']= df['Discount'].fillna(0)
```

```python
df['NetPrice'] = df['Price'] * (1 - df['Discount'] / 100)
```

```python
df['TotalSale'] = df['Quantity'] * df['NetPrice']
```

```python
# Calculate the average TotalSale for each store
store_avg_sales = df.groupby('StoreID')['TotalSale'].mean()

# Find the store with the highest average TotalSale
highest_average_TotalSale_for_each_store = store_avg_sales.idxmax()

# Create a DataFrame containing the highest average TotalSale for each store
highest_average_TotalSale_for_each_store = df[df['StoreID'] == highest_average_TotalSale_for_each_store]


summary_df = highest_average_TotalSale_for_each_store.rename(columns={'TotalSale': 'AvgTotalSale'}).reset_index(drop=True)
print(summary_df)
```
